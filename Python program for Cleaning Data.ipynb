import the csv file 2. weatherHistory -Inserted data through Python into Google Colab

```python
import pandas as pd
```

```python
df=pd.read_csv('/content/2. weatherHistory -Inserted data through Python.csv')
df
```

```python
df=df.round(2)
df
```

```python
import numpy as np
df=df.replace(0,value=np.nan)
df
```

```python
df.drop(columns=['Formatted Date'],inplace=True)
df
```

```python
df.drop(columns=['Loud Cover'],inplace=True)
df
```

```python
df.replace('null',pd.NA, inplace=True)
df
```

```python
df=df.dropna(subset=['Precip Type'])
df
```

```python
df=df.fillna(method='ffill')
df
```

```python
df.to_csv('cleaned_file.csv',index=False)
```
